```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats
import matplotlib
```

```python
plt.rcParams['font.sans-serif'] = 'Times New Roman'
```

```python
matplotlib.rcParams.update({'font.size': 10.2})
with open("cnn.log", 'r') as f:
    lines = f.readlines()
lines = lines[1:]
for i in range(len(lines)):
    lines[i] = lines[i].split(',')
for i in range(len(lines)):
    for j in range(len(lines[i])):
        lines[i][j] = float(lines[i][j].strip(', \n'))
loss = [[],[],[],[]]
acc = [[],[],[],[]]
for i in range(min(len(lines), 50)):
    for j in range(4):
        loss[j].append(lines[i][2*j])
    for j in range(4):
        acc[j].append(lines[i][2*j+1] * 100)

fig = plt.figure(figsize=(8,4))
names = ['Paper(Train)', 'iPad(Valid)', 'Phone in Hand(Valid)', 'Phone on Table(Valid)']
colors = ['#376DF9', '#36D6C6', '#FF440B', '#D4D247']
plt.subplot(121)
for i in [0, 1, 3, 2]:
    plt.semilogy(loss[i], label = names[i], color=colors[i], linewidth=1.2)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(122)
for i in [0, 1, 3, 2]:
    plt.semilogy(acc[i], label = names[i], color=colors[i], linewidth=1.2)
plt.xlabel('Epoch')
plt.ylabel('Acc/%')
plt.legend()

plt.tight_layout()
plt.show()
fig.savefig('nn-1.pdf', bbox_inches='tight')
```

```python
loss = [[],[]]
acc = [[],[]]
matplotlib.rcParams.update({'font.size': 12})
with open("snn2.log", 'r') as f:
    lines = f.readlines()
lines = lines[1:]
for i in range(len(lines)):
    lines[i] = lines[i].split(',')
for i in range(len(lines)):
    for j in range(len(lines[i])):
        lines[i][j] = float(lines[i][j].strip(', \n'))
for i in range(len(lines)):
    loss[0].append(lines[i][0])
    acc[0].append(lines[i][1] * 100)
    loss[1].append(lines[i][2])
    acc[1].append(lines[i][3] * 100)

fig = plt.figure(figsize=(8,4))
names = ['Train', 'Valid']
colors = ['#FF440B', '#376DF9']

plt.subplot(121)
for i in range(2):
    plt.semilogy(loss[i], label = names[i], color=colors[i], linewidth=1.2)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(122)
for i in range(2):
    plt.plot(acc[i], label = names[i], color=colors[i],linewidth=1.2)
plt.xlabel('Epoch')
plt.ylabel('Acc/%')
plt.legend()

plt.tight_layout()
plt.show()
fig.savefig('nn-2.pdf', bbox_inches='tight')
```
